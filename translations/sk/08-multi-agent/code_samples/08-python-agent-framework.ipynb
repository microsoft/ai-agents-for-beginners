## Nastavenie


````python
import logging
logging.getLogger("agent_framework.azure").setLevel(logging.ERROR)

%pip install agent-framework azure-ai-projects azure-identity --quiet

import os
import asyncio

from agent_framework import AgentResponseUpdate, WorkflowBuilder
from agent_framework.azure import AzureAIProjectAgentProvider
from azure.identity import AzureCliCredential
````

````python
provider = AzureAIProjectAgentProvider(credential=AzureCliCredential())
````

## Prečo multi-agentné systémy?

Úlohy zo skutočného sveta, ako je plánovanie cesty, vyžadujú rôzne druhy odborných znalostí — logistiku, miestne znalosti, rozpočtovanie a ďalšie. Jeden agent, ktorý sa snaží zvládnuť všetko, sa rýchlo stáva neprehľadným.

Multi-agentné systémy to riešia prostredníctvom **špecializácie**: každý agent sa zameriava na jednu oblasť odbornosti, čo prináša výsledky vyššej kvality než všeobecný agent. Tiež zlepšujú **škálovateľnosť** — môžete pridať nových agentov (napr. špecialistu na lety, kritika reštaurácií) bez prepisovania existujúceho pracovného postupu. Agenti spolupracujú cez štruktúrovaný tok, kde si odovzdávajú kontext jeden druhému.


## Vytváranie špecializovaných agentov


````python
planner_agent = await provider.create_agent(
    name="TravelPlanner",
    instructions="You are a travel planning specialist. Create detailed trip itineraries based on the traveler's preferences. Include daily schedules, must-see attractions, and logistical tips.",
)

concierge_agent = await provider.create_agent(
    name="TravelConcierge",
    instructions="You are a travel concierge who reviews and enhances trip plans. Review the plan for completeness, add local insider tips, suggest restaurants, and identify potential issues. Provide your feedback in a constructive format.",
)
````

## Vytváranie sekvenčného pracovného postupu

`WorkflowBuilder` vám umožňuje prepojiť agentov do usmerneného grafu. Tu vytvárame jednoduchý dvojkrokový proces: **TravelPlanner** zostaví itinerár, potom ho **TravelConcierge** skontroluje a vylepší.


````python
workflow = WorkflowBuilder(start_executor=planner_agent) \
    .add_edge(planner_agent, concierge_agent) \
    .build()

last_author = None
events = workflow.run("Plan a 5-day trip to Paris for a food-loving couple on a $3000 budget.", stream=True)
async for event in events:
    if event.type == "output" and isinstance(event.data, AgentResponseUpdate):
        update = event.data
        author = update.author_name
        if author != last_author:
            if last_author is not None:
                print()
            print(f"\n{'='*50}")
            print(f"🤖 {author}:")
            print(f"{'='*50}")
            last_author = author
        print(update.text, end="", flush=True)
````

## Pridanie ďalších agentov do pracovného postupu

Jednou z najväčších výhod viacagentného vzoru je jeho jednoduchá rozšíriteľnosť. Nižšie pridávame agenta **BudgetReviewer**, ktorý kontroluje plán voči rozpočtu cestovateľa, označuje položky, ktoré by mohli prekročiť limit nákladov, a navrhuje alternatívy na úsporu peňazí. Pracovný postup teraz postupne spúšťa troch agentov:

```
TravelPlanner → TravelConcierge → BudgetReviewer
```


````python
budget_agent = await provider.create_agent(
    name="BudgetReviewer",
    instructions="You are a budget-conscious travel advisor. Review the proposed trip plan and concierge enhancements against the traveler's stated budget. Estimate costs for flights, hotels, meals, and activities. Flag anything that risks exceeding the budget and suggest cost-saving alternatives while preserving the trip's quality.",
)

extended_workflow = WorkflowBuilder(start_executor=planner_agent) \
    .add_edge(planner_agent, concierge_agent) \
    .add_edge(concierge_agent, budget_agent) \
    .build()

last_author = None
events = extended_workflow.run("Plan a 5-day trip to Paris for a food-loving couple on a $3000 budget.", stream=True)
async for event in events:
    if event.type == "output" and isinstance(event.data, AgentResponseUpdate):
        update = event.data
        author = update.author_name
        if author != last_author:
            if last_author is not None:
                print()
            print(f"\n{'='*50}")
            print(f"🤖 {author}:")
            print(f"{'='*50}")
            last_author = author
        print(update.text, end="", flush=True)
````

## Zhrnutie

V tejto lekcii ste sa naučili:

1. **Vytvárať špecializovaných agentov** — každý s jasnou úlohou (plánovanie, recepcia, kontrola rozpočtu).
2. **Prepojiť agentov do sekvenčného pracovného toku** pomocou `WorkflowBuilder` a `add_edge`.
3. **Streamovať výstup** z viac-agentového potrubia a sledovať, ktorý agent práve hovorí.
4. **Rozšíriť pracovný tok** pridaním nových agentov do reťazca bez úprav existujúcich.

Dizajnový vzor multi-agentov zachováva každého agenta jednoduchým, pričom vytvára bohatšie a dôkladnejšie prekontrolované výsledky, než akýkoľvek jednotný agent dokáže sám dosiahnuť.
