## രൂപരേഖ

1. സജ്ജീകരണം

```python
# Setup cell: keep it short and deterministic
from __future__ import annotations

import math
import random

SEED = 21
random.seed(SEED)
SEED
```

## പടി 1 - ഒരു ചെറിയ ഉദാഹരണത്തോടെ തുടങ്ങുക

എളുപ്പത്തിലുള്ള ഭാഷയിൽ അടുത്ത സെൽ എന്താണ് ചെയ്യുന്നത് വിശദീകരിക്കുക.


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```

## അഭ്യാസങ്ങൾ

- ഒരു വ്യത്യസ്ത ഇൻപുട്ട് പരീക്ഷിക്കുക.
- കോഡ് പ്രവർത്തിപ്പിക്കുന്നതിന് മുമ്പ് ഔട്ട്പുട്ട് പ്രവചിക്കുക.
- ഒരു സാധാരണ പിശകും അത് എങ്ങനെ പരിഹരിക്കാമെന്നുമൊന്ന് രേഖപ്പെടുത്തുക.


```python
# Exercise answer scaffold
def describe(values: list[float]) -> dict[str, float]:
    return {"min": min(values), "max": max(values)}

describe(sines)
```
