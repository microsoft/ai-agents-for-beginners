## ಹಂತ 1 - ಸಣ್ಣ ಉದಾಹರಣೆಯಿಂದ ಪ್ರಾರಂಭಿಸಿ

ಮುಂದಿನ ಸೆಲ್ ಯಾವ ಕೆಲಸ ಮಾಡುತ್ತದೆ ಎಂಬುದನ್ನು ಸರಳ ಭಾಷೆಯಲ್ಲಿ ವಿವರಿಸಿ.


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```

## ಅಭ್ಯಾಸಗಳು

- ಇನ್ನೊಂದು ಇನ್ಪುಟ್ ಪ್ರಯತ್ನಿಸಿ.
- ಕೋಡ್ ಚಲಿಸುವ ಮೊದಲು ಫಲಿತಾಂಶವನ್ನು ಊಹಿಸಿ.
- ಒಂದು ಸಾಮಾನ್ಯ ತಪ್ಪನ್ನು ಗಮನಿಸಿ ಮತ್ತು ಅದನ್ನು ಹೇಗೆ ಸರಿಪಡಿಸಬೇಕು ಎಂಬುದನ್ನು ತಿಳಿಸಿ.


```python
# Exercise answer scaffold
def describe(values: list[float]) -> dict[str, float]:
    return {"min": min(values), "max": max(values)}

describe(sines)
```
