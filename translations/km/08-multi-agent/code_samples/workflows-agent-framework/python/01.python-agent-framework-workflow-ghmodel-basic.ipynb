бһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһ“бҹҒбҹҮбһҺбҹӮбһ“бһ¶бҹҶбһўбҹҶбһ–бһёбһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһҠбҹҸбһҳбһ¶бһ“бһўбҹҶбһҺбһ¶бһ… **Workflow Builder** бһҡбһ”бһҹбҹӢ Microsoft Agent Frameworkбҹ” бһҡбҹҖбһ“бһ–бһёбһҡбһ”бҹҖбһ”бһ”бһ„бҹ’бһҖбһҫбһҸбһ“бһ·бһ„бһҡбҹҖбһ”бһ…бҹҶбһ…бһӣбҹҗбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҮбҹҶбһ бһ¶бһ“ бһҠбҹӮбһӣбһўбһ¶бһ…бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһўбһ¶бһҮбһёбһңбһҖбһҳбҹ’бһҳбһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһү бһ“бһ·бһ„ Ш§Щ„ШӘЩҶШіЩҠЩӮ Ш№Щ…Щ„ЩҠШ§ШӘ Ш§Щ„Ш°ЩғШ§ШЎ Ш§Щ„Ш§ШөШ·ЩҶШ§Ш№ЩҠ ШЁШЈЩғШ«Шұ Щ…ЩҶ ШіЩ„Ш§ШіШ©бҹ”

## рҹҺҜ бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶

### рҹҸ—пёҸ **бһҹбҹ’бһҗбһ¶бһ”бһҸбҹ’бһҷбһҖбһҳбҹ’бһҳбһ…бһӣбҹҗбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё**

- **бһҖбһ¶бһҡбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҠбҹ„бһҷбһ”бһҺбҹ’бһҠбһ¶бһӣбһұбҹ’бһҷбһҳбһ¶бһ“бһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚ**: бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҺбҹҚбһ…бһӣбҹҗбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһ“бһ·бһ„бһҖбһ¶бһҡбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ
- **бһҡбһ…бһ“бһ¶бһ…бһӣбҹҗбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҸбһ¶бһҳбһҡбһјбһ”бһ—бһ¶бһ–**: бһ”бһ„бҹ’бһҖбһҫбһҸ бһ“бһ·бһ„бһҳбһҫбһӣбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ…бһӣбҹҗбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё
- **бһҖбһ¶бһҡбһҸбҹҶбһҺбһ¶бһҖбҹӢбһ‘бҹҶбһ“бһ„бһӮбҹҶбһҡбһј GitHub**: бһ”бҹ’бһҡбһҫбһӮбҹҶбһҡбһј AI бһҖбҹ’бһ“бһ»бһ„бһ”бһҡбһ·бһ”бһ‘бһ…бһӣбҹҗбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё

### рҹ”„ **бһҖбһ¶бһҡбһҡбҹҖбһ”бһ…бҹҶбһ“бһёбһҸбһ·бһңбһ·бһ’бһё**
- **бһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҸбһ¶бһҳбһӣбҹҶбһҠбһ¶бһ”бҹӢ**: бһҠбһ”бһ“бҹӢбһ—бһ¶бһҡбһҖбһ·бһ…бҹ’бһ…бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҸбһ¶бһҳбһӣбҹҶбһҠбһ¶бһ”бҹӢбһ  Р»РҫРіРёСҮРҪСӢР№

- **бһҖбһ¶бһҡбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҖбҹҶбһ бһ»бһҹ**: бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһҹбҹ’бһҠбһ¶бһҡбһЎбһҫбһ„бһңбһ·бһүбһҖбҹҶбһ бһ»бһҹ бһ“бһ·бһ„бһ—бһ¶бһ–бһҡбһ№бһ„бһҳбһ¶бҹҶбһ“бҹғбһ…бһӣбҹҗбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё

### рҹ“Ҡ **бһӮбҹҶбһҡбһјбһ…бһӣбҹҗбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҹбһ бһӮбҹ’бһҡбһ¶бһҹ**
- **бһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһўбһ¶бһҮбһёбһңбһҖбһҳбҹ’бһҳ**: бһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбһҠбҹ„бһҷбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһүбһҡбһ”бһҹбҹӢбһўбһ„бҹ’бһӮбһ—бһ¶бһ–
- **бһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһ”бһҹбһҳбҹ’бһҡбһҪбһӣбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҳбһ бһ¶бһҹбһ¶бһӣ**: бһҹбһҳбҹ’бһҡбһ”бһҹбһҳбҹ’бһҡбһҪбһӣбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһ–бһ·бһҹбҹҒбһҹбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“

````python
# Core components for building sophisticated agent workflows
from agent_framework import WorkflowBuilder, WorkflowEvent, WorkflowViz
from agent_framework.azure import AzureAIProjectAgentProvider
from azure.identity import AzureCliCredential
````

````python
# рҹ“Ұ Import Environment and System Utilities
# Essential libraries for configuration and environment management

import os                      # рҹ”§ Environment variable access
from dotenv import load_dotenv # рҹ“Ғ Secure configuration loading
````

````python
# рҹ”§ Initialize Environment Configuration
# Load GitHub Models API credentials from .env file
load_dotenv()
````

````python
# Configure the Azure AI Foundry provider with keyless authentication
provider = AzureAIProjectAgentProvider(credential=AzureCliCredential())
````

````python
REVIEWER_NAME = "Concierge"
REVIEWER_INSTRUCTIONS = """
    You are an are hotel concierge who has opinions about providing the most local and authentic experiences for travelers.
    The goal is to determine if the front desk travel agent has recommended the best non-touristy experience for a traveler.
    If so, state that it is approved.
    If not, provide insight on how to refine the recommendation without using a specific example. 
    """
````

````python
FRONTDESK_NAME = "FrontDesk"
FRONTDESK_INSTRUCTIONS = """
    You are a Front Desk Travel Agent with ten years of experience and are known for brevity as you deal with many customers.
    The goal is to provide the best activities and locations for a traveler to visit.
    Only provide a single recommendation per response.
    You're laser focused on the goal at hand.
    Don't waste time with chit chat.
    Consider suggestions when refining an idea.
    """
````

````python
reviewer_agent = await provider.create_agent(
    name=REVIEWER_NAME,
    instructions=REVIEWER_INSTRUCTIONS,
)

front_desk_agent = await provider.create_agent(
    name=FRONTDESK_NAME,
    instructions=FRONTDESK_INSTRUCTIONS,
)
````

````python
workflow = (
    WorkflowBuilder(start_executor=front_desk_agent)
    .add_edge(front_desk_agent, reviewer_agent)
    .build()
)
````

````python
print("Generating workflow visualization...")
viz = WorkflowViz(workflow)
# Print out the mermaid string.
print("Mermaid string: \n=======")
print(viz.to_mermaid())
print("=======")
# Print out the DiGraph string.
print("DiGraph string: \n=======")
print(viz.to_digraph())
print("=======")
# SVG export needs the optional graphviz extra plus the graphviz system binary;
# fall back gracefully if it is not available.
try:
    svg_file = viz.export(format="svg")
    print(f"SVG file saved to: {svg_file}")
except ImportError as e:
    svg_file = None
    print(f"SVG export skipped (install graphviz to enable): {e}")
````

````python
# Display the exported workflow SVG inline in the notebook

from IPython.display import SVG, display, HTML
import os

print(f"Attempting to display SVG file at: {svg_file}")

if svg_file and os.path.exists(svg_file):
    try:
        # Preferred: direct SVG rendering
        display(SVG(filename=svg_file))
    except Exception as e:
        print(f"вҡ пёҸ Direct SVG render failed: {e}. Falling back to raw HTML.")
        try:
            with open(svg_file, "r", encoding="utf-8") as f:
                svg_text = f.read()
            display(HTML(svg_text))
        except Exception as inner:
            print(f"вқҢ Fallback HTML render also failed: {inner}")
else:
    print("вқҢ SVG file not found. Ensure viz.export(format='svg') ran successfully.")
````

````python
# Workflow.run_stream is no longer part of the public API; the current Workflow
# returns a results object whose `get_outputs()` produces the AgentResponse from
# each output executor. The reviewer (last stage) is the only output here.
events = await workflow.run("I would like to go to Paris.")
outputs = events.get_outputs()
result = outputs[0].text if outputs else ""
````
