# 🔍 Microsoft Foundry (.NET) తో ఎంటర్ప్రైజ్ RAG

## 📋 అభ్యాస లక్ష్యాలు

ఈ నోట్బుక్ Microsoft Foundryతో .NETలో Microsoft Agent Framework ఉపయోగించి ఎంటర్ప్రైజ్-తరగతి Retrieval-Augmented Generation (RAG) సిస్టమ్స్‌ను నిర్మించడం ఎలా అనేదాన్ని చూపిస్తుంది. మీరు డాక్యుమెంట్లలో శోధించి ఎంటర్ప్రైజ్ సెక్యూరిటీ మరియు స్కేలబిలిటీతో ఖచ్చితమైన, సందర్భానుకూల ప్రతిస్పందనలు ఇవ్వగల ప్రొడక్షన్-రెడీ ఏజెంట్లను ఎలా సృష్టించాలో నేర్చుకుంటారు.

**మీరు నిర్మించబోయే ఎంటర్ప్రైజ్ RAG సామర్థ్యాలు:**
- 📚 **డాక్యుమెంట్ ఇంటెలిజెన్స్**: Azure AI సేవలతో ఆధునిక డాక్యుమెంట్ ప్రాసెసింగ్
- 🔍 **సెమాంటిక్ శోధన**: ఎంటర్ప్రైజ్ ఫీచర్లతో అత్యుత్తమ పనితీరున్న వెక్టర్ శోధన
- 🛡️ **సెక్యూరిటీ ఇంటిగ్రేషన్**: పాత్ర-ఆధారిత యాక్సెస్ మరియు డేటా రక్షణ ప్యాటర్న్లు
- 🏢 **స్కేలబుల్ ఆర్కిటెక్చర్**: మానిటరింగ్‌తో ప్రొడక్షన్-రెడీ RAG సిస్టములు

## 🎯 ఎంటర్ప్రైజ్ RAG వాస్తవిక నిర్మాణం

### ముఖ్య ఎంటర్ప్రైజ్ భాగాలు
- **Microsoft Foundry**: సెక్యూరిటీ మరియు కంప్లయన్స్‌తో నిర్వహించే ఎంటర్ప్రైజ్ AI ప్లాట్‌ఫారమ్
- **Persistent Agents**: సంభాషణ చరిత్ర మరియు సందర్భ నిర్వహణతో స్టేట్‌ఫుల్ ఏజెంట్లు
- **Vector Store Management**: డాక్యుమెంట్ సూచికనిర్మాణం మరియు పునఃప్రాప్తి కోసం ఎంటర్ప్రైజ్-స్థాయి నిర్వహణ
- **Identity Integration**: Azure AD ప్రామాణీకరణ మరియు పాత్ర-ఆధారిత యాక్సెస్ నియంత్రణ

### .NET ఎంటర్ప్రైజ్ లాభాలు
- **Type Safety**: RAG ఆపరేషన్లు మరియు డేటా స్ట్రక్చర్‌ల కోసం కాంపైల్-టైమ్ ధృవీకరణ
- **Async Performance**: అడ్డంకిలేని డాక్యుమెంట్ ప్రాసెసింగ్ మరియు శోధన ఆపరేషన్లు
- **Memory Management**: భారీ డాక్యుమెంట్ సేకరణల కోసం సమర్థవంతమైన వనరు వినియోగం
- **Integration Patterns**: డిపెండెన్సీ ఇంజెక్షన్‌తో స్థానిక Azure సేవల సమీకరణ

## 📊 Enterprise RAG Patterns

### Document Management Patterns
- **Bulk Upload**: పెద్ద డాక్యుమెంట్ సేకరణల సమర్థవంతమైన ప్రాసెసింగ్
- **Incremental Updates**: రియల్-టైమ్ డాక్యుమెంట్ జోడింపు మరియు సవరణ
- **Version Control**: డాక్యుమెంట్ వెర్షనింగ్ మరియు మార్పుల ట్రాకింగ్
- **Metadata Management**: సంపన్న డాక్యుమెంట్ లక్షణాలు మరియు వర్గీకరణ

### Search & Retrieval Patterns
- **Hybrid Search**: ఉత్తమ ఫలితాల కోసం సెమాంటిక్ మరియు కీవర్డ్ శోధనలను కలపడం
- **Faceted Search**: బహుముఖ ఫిల్టరింగ్ మరియు వర్గీకరణ
- **Relevance Tuning**: డొమైన్-స్పెసిఫిక్ అవసరాలకు అనుగుణంగా కస్టమ్ స్కోరింగ్ ఆల్గోరిథమ్స్
- **Result Ranking**: బిజినెస్ లాజిక్ సమీకరణతో ఆధునిక ర్యాంకింగ్

### Security Patterns
- **Document-Level Security**: ప్రతి డాక్యుమెంట్ కోసం సూక్ష్మ-గ్రేడ్ యాక్సెస్ నియంత్రణ
- **Data Classification**: స్వయంచాలక సున్నితత్వ లేబెలింగ్ మరియు రక్షణ
- **Audit Trails**: అన్ని RAG ఆపరేషన్ల సమగ్ర లాగింగ్
- **Privacy Protection**: PII గుర్తింపు మరియు రిడాక్షన్ సామర్థ్యాలు

### Data Protection

- **Access Controls**: వినియోగదారు మరియు గ్రూప్ అనుమతుల కోసం Azure ADతో సమీకరణ
- **Data Residency**: కంప్లయన్స్ కోసం భౌగోళిక డేటా స్థాన నియంత్రణలు
- **Backup & Recovery**: ఆటోమేటెడ్ బ్యాకప్ మరియు విపత్తు పునరుద్ధరణ సామర్థ్యాలు

- **Resource Pooling**: ఖర్చైన వనరుల సమర్థవంతమైన పునర్వినియోగం
- **Garbage Collection**: మెమరీ కేటాయింపు నమూనాలను ఆప్టిమైజ్ చేయడం
- **Connection Management**: సరైన Azure సేవ కనెక్షన్ జీవితచక్రం

### క్యాశింగ్ వ్యూహాలు
- **Query Caching**: తరచుగా నడిపే శోధనలను క్యాష్ చేయండి
- **Document Caching**: హాట్ డాక్యుమెంట్ల కోసం ఇన్-మెమరీ క్యాషింగ్
- **Index Caching**: వెక్టర్ ఇండెక్స్ క్యాషింగ్ ఆప్టిమైజేషన్
- **Result Caching**: ఉత్పత్తి చేసిన ప్రతిస్పందనల బుద్ధిమంతమైన క్యాషింగ్

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-4.1-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
